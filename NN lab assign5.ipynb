```python
#This is the python code for Neural Network assignment 5

import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid function
def sigmoid_derivative(x):
    return x * (1 - x)

# Input data
input_data = np.array([[0.6, 0.8, 1]])  # Adding bias term
# Target output
target_output = np.array([[0.9]])

# Learning rate
alpha = 0.3

# Number of iterations
epochs = 10000

# Number of input, hidden, and output nodes
input_nodes = input_data.shape[1]
hidden_nodes = 4
output_nodes = 1

# Initialize weights with random values, including bias term
weights_input_hidden = np.random.uniform(size=(input_nodes, hidden_nodes))
weights_hidden_output = np.random.uniform(size=(hidden_nodes, output_nodes))

# Training the neural network
for epoch in range(epochs):
    # Forward propagation
    hidden_input = np.dot(input_data, weights_input_hidden)
    hidden_output = sigmoid(hidden_input)

    output_input = np.dot(hidden_output, weights_hidden_output)
    predicted_output = sigmoid(output_input)

    # Backpropagation
    # Calculate the error
    error = target_output - predicted_output

    # Calculate gradients
    output_error = error * sigmoid_derivative(predicted_output)

    hidden_layer_error = output_error.dot(weights_hidden_output.T) * sigmoid_derivative(hidden_output)

    # Update weights using gradients and learning rate
    weights_hidden_output += hidden_output.T.dot(output_error) * alpha
    weights_input_hidden += input_data.T.dot(hidden_layer_error) * alpha

# Print the final predicted output
print("Final Predicted Output after Training:")
print(predicted_output)
```

```python
#Kolmogorov Theorem



import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Generate random input data
np.random.seed(0)
X = np.random.rand(100, 1) * 2 * np.pi
Y = np.sin(X)

# Build a neural network with one hidden layer
model = Sequential()
model.add(Dense(10, input_dim=1, activation='relu'))  # 10 neurons in the hidden layer
model.add(Dense(1, activation='linear'))  # Output layer with linear activation function

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(X, Y, epochs=1000, batch_size=10, verbose=0)

# Test the model
test_input = np.array([[1.5]])  # Test input value
predicted_output = model.predict(test_input)
print("Predicted Output for Input 1.5:", predicted_output[0, 0])

# Output: Predicted output for input 1.5 using the trained neural network


```
